## Section 6 — While Loop

```python
stock = 10

# 1. Print "Item sold — stock: X" and reduce stock by 1
#    until stock reaches 0
# Expected output:
# Item sold — stock: 10
# Item sold — stock: 9
# ...
# Item sold — stock: 1

stock = 10

while stock > 0:
    print(f"Item sold — stock: {stock}")
    stock=stock-1
```

```python
attempts = 0

# 2. Print "Login attempt X" and increase attempts by 1
#    Stop after 3 attempts
# Expected output:
# Login attempt 1
# Login attempt 2
# Login attempt 3


attempts = 0

while attempts < 3:
    attempts += 1
    print(f"Login attempt {attempts}")
```

```python
# 3. Keep asking user for payment status using input()
#    If "paid" — print "Payment confirmed" and exit
#    Otherwise print "Try again"

while True:
    status = input("Enter payment status: ")

    if status.lower() == "paid":
        print("Payment confirmed")
        break
    else:
        print("Try again")
```

```python
balance = 1000
withdrawal = 200

# 4. Keep withdrawing 200 from balance
#    Print balance after each withdrawal
#    Stop when balance reaches 0
# Expected output:
# Balance after withdrawal: 800
# Balance after withdrawal: 600
# Balance after withdrawal: 400
# Balance after withdrawal: 200
# Balance after withdrawal: 0

balance = 1000
withdrawal = 200

while balance > 0:
   print(f"Balance after withdrawal: {balance}")
   balance -= withdrawal
```

## Section 7 — break

```python
products = ["Mobile", "Laptop", "Tablet", "Smartwatch", "Earbuds"]

# 1. Stop and print "Found it" when you reach "Tablet"
#    Print "Checking: X" for everything before it
# Expected output:
# Checking: Mobile
# Checking: Laptop
# Found it

products = ["Mobile", "Laptop", "Tablet", "Smartwatch", "Earbuds"]

for p in products:
    if p == "Tablet":
        print("Found it")
        break
    print(f"Checking: {p}")
```

```python
order_amounts = [1500, 800, 2300, 450, 1100]

# 2. Loop through amounts
#    Stop the loop when you find an amount below 500
#    Print each amount before stopping
# Expected output:
# Processing: Rs.1500
# Processing: Rs.800
# Processing: Rs.2300
# Found amount below 500: Rs.450 — stopping

order_amounts = [1500, 800, 2300, 450, 1100]

for a in order_amounts:
    if a < 500:
        print(f"Found amount below 500: Rs.{a} — stopping")
        break
    print(f"Processing: Rs.{a}")
```

```python
# 3. Keep asking user to enter an order ID using input()
#    If they type "quit" — print "Exiting" and stop
#    Otherwise print "Order received: X"

while True:
    id = input("Enter order ID: ")

    if id.lower() == "quit":
        print("Exiting")
        break
    print(f"Order received: {id}")
```

```python
order_amounts = [1500, 0, 800, 0, 2300, 0, 1100]

# 1. Skip zero amounts, print "Processing: Rs.X" for valid amounts
# Expected output:
# Processing: Rs.1500
# Processing: Rs.800
# Processing: Rs.2300
# Processing: Rs.1100

order_amounts = [1500, 0, 800, 0, 2300, 0, 1100]

for d in order_amounts:
    if d==0:
        continue

    print(f"Processing: Rs.{d}")
```

## Section 10 — Mixed (Stretch)

```python
orders = [
    {"id": "ORD001", "status": "delivered", "amount": 1500},
    {"id": "ORD002", "status": "pending",   "amount": 0},
    {"id": "ORD003", "status": "cancelled", "amount": 800},
    {"id": "ORD004", "status": "delivered", "amount": 2300},
    {"id": "ORD005", "status": "delivered", "amount": 0}
]

# 1. Skip orders with amount 0 using continue
#    Stop at cancelled using break
#    Print valid orders: "ORD001 — delivered — Rs.1500"
# Expected output:
# ORD001 — delivered — Rs.1500

for w in orders:
    if w["amount"] == 0:
        continue
    if w["status"] == "cancelled":
        break
    print(f"{w['id']} — {w['status']} — Rs.{w['amount']}")
```

```python
products = [
    {"name": "Mobile",     "stock": 5,  "price": 25000},
    {"name": "Laptop",     "stock": 0,  "price": 85000},
    {"name": "Tablet",     "stock": 3,  "price": 45000},
    {"name": "Earbuds",    "stock": 0,  "price": 5000},
    {"name": "Smartwatch", "stock": 8,  "price": 15000}
]

# 2. Skip products with stock 0 using continue
#    For remaining products print: "Mobile — Rs.25000 — 5 in stock"
# Expected output:
# Mobile — Rs.25000 — 5 in stock
# Tablet — Rs.45000 — 3 in stock
# Smartwatch — Rs.15000 — 8 in stock

for pro in products:
    if pro["stock"] == 0:
        continue
    print(f"{pro['name']} — Rs.{pro['price']} — {pro['stock']} in stock")
```
