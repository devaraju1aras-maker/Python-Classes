```python
# Dictionaries

# Value are stored as key:value pairs
# ordered
# Duplicate key are not allowed
# Heterogenous
# Mutable
# keys must be immutable elements, values can be mutable elements
```

What is allowed as a key?

| Type | Allowed as Key | Reason |
|---|---|---|
| string | Yes | Fixed, cannot change |
| int, float | Yes | Fixed, cannot change |
| tuple | Yes | Fixed, cannot change |
| list | No | Can change after creation |
| dict | No | Can change after creation |

What is allowed as a value?

| Type | Allowed as Value |
|---|---|
| string | Yes |
| int, float | Yes |
| list | Yes |
| tuple | Yes |
| dict | Yes |
| set | Yes |
| bool | Yes |

Methods for dictionary -
| Method | What it does |
|---|---|
| .get() | Safe access by key — no error if missing |
| .keys() | Returns all keys |
| .values() | Returns all values |
| .items() | Returns all key-value pairs |
| .update() | Updates one or more keys at once |
| .pop() | Removes a key and returns its value |
| .copy() | Returns a copy of the dictionary |
| .clear() | Empties the dictionary |

Return types for methods- 

| Method | Returns |
|---|---|
| .get() | Value, None, or default |
| .keys() | dict_keys view |
| .values() | dict_values view |
| .items() | dict_items view |
| .update() | None |
| .pop() | Removed value |
| .copy() | New dictionary |
| .clear() | None |

```python
# Access values 

# Access single values
Fruits={"Apple":"Red","Banana":"Yellow","orange":"orange","Grapes":"Green"}
print(Fruits["Apple"])

# Key and values as different datatypes - heterogenous property
Fruits={1:"Red",2:"Yellow",3:"orange",4:"Green"}
print(Fruits[1])

# heterogenous property
Fruits={"Apple":1,"Banana":2,"orange":3,"Grapes":4}
print([Fruits["orange"]])

# List, sets, dicts, are mutable and cannot be stored as keys but tuples can be
Fruits={1:"Red",1.5:"Yellow",True:"orange","Grapes":"Green",(1,2,3):"Rainbow"}
print(Fruits[(1,2,3)])

# Duplicate key behavior
Fruits={"Apple":1,"Apple":1.5,"orange":True,"Grapes":"Green"} # duplicate key behavior
print(Fruits["Apple"])
print(Fruits)
```

```python
# get method
Fruits={"Apple":"Red","Banana":"Yellow","orange":"orange","Grapes":"Green"}
print(Fruits.get("Apple")) # Access a value

print(Fruits.get("Pear"))

#safe method to access a value
# get method does not throw a key error if the value is not present

```

```python
# Keys, values, items

print(Fruits.keys())

print(Fruits.values())

print(Fruits.items())

#you can convert keys, values to any other datatype like list, sets or tuples using type conversion
l1=list(Fruits.keys())
print(l1)

l2=list(Fruits.values())
print(l2)

l3=list(Fruits.items())
print(l3)

t1=tuple(Fruits.keys())
print(t1)

s1=set(Fruits.values())
print(s1)
```

```python
#Update,pop,copy,clear

Fruits={"Apple":"Red","Banana":"Yellow","orange":"orange","Grapes":"Green"}
Fruits["Apple"]="Blue" # update a value through key
print(Fruits)

Fruits.update({"Apple":"Pink"}) # update single element
print(Fruits)

Fruits.update({"Apple":"Pink","Banana":"Blue"}) #update multiple elements
print(Fruits)

Fruits["Pear"]="Green"
print(Fruits)

Fruits.update({"Papaya":"Yellow","Mango":"Green"}) # update multiple elements
print(Fruits)

# if a key is already present it will update it with new value
# if key is not present it will add it as a new item.
```

```python
#Remove - POP Method
Fruits={"Apple":"Red","Banana":"Yellow","orange":"orange","Grapes":"Green"}

Fruits.pop("Apple")
print(Fruits)

Fruits.pop("Peach","Not Found") # custom message if key not found (do not give print)
# without custom message pop will thorow an error if key not found.
```

```python
# Copy and Clear

#Copy

Fruits={"Apple":"Red","Banana":"Yellow","orange":"orange","Grapes":"Green"}
backup=Fruits.copy()
print(Fruits)

#Clear - Clear the entire dictionary

Fruits.clear()
print(Fruits)
```

```python
# nested structures

l1=[1,2,3,[1,2,3]]
print(l1[0])
print(l1[3])
print(l1[3][0])

# list inside dictionary as value
dl={"Apple":["Red","Green"],"Capsicum":["Yellow","Green","Red"],"Marker":["Red","Blue","Green"]}
print(dl["Capsicum"][1])

#tuples inside dictionary as values
d2={"Apple":("Red","Green"),"Capsicum":("Yellow","Green","Red"),"Marker":("Red","Blue",("Green"))}
print(d2["Capsicum"][2])

#Sets inside dictionary as values
d2={"Apple":("Red","Green"),"Capsicum":("Yellow","Green","Red"),"Marker":("Red","Blue",("Green"))}
print(d2["Capsicum"])

#Dictionaries inside dictionaries as values
d2={"Apple":{"Red":2,"Green":3},"Capsicum":{"Yellow":2,"Green":4,"Red":5},"Marker":{"Red":4,"Blue":3,"Green":2}}

print(d2["Capsicum"]["Yellow"])
```

# String Methods — Complete Reference

| Method | Syntax | Required Args | Optional Args | Returns |
|---|---|---|---|---|
| .upper() | str.upper() | None | None | New string |
| .lower() | str.lower() | None | None | New string |
| .title() | str.title() | None | None | New string |
| .isupper() | str.isupper() | None | None | Boolean |
| .islower() | str.islower() | None | None | Boolean |
| .isdigit() | str.isdigit() | None | None | Boolean |
| .isalpha() | str.isalpha() | None | None | Boolean |
| .strip() | str.strip(chars='') | None | chars | New string |
| .lstrip() | str.lstrip(chars='') | None | chars | New string |
| .rstrip() | str.rstrip(chars='') | None | chars | New string |
| .replace() | str.replace(old, new, count=-1) | old, new | count | New string |
| .split() | str.split(sep=None, maxsplit=-1) | None | sep, maxsplit | List |
| .join() | str.join(iterable) | iterable | None | New string |
| .find() | str.find(sub, start=0, end=len(str)) | sub | start, end | Integer or -1 |
| .index() | str.index(sub, start=0, end=len(str)) | sub | start, end | Integer or ValueError |
| .count() | str.count(sub, start=0, end=len(str)) | sub | start, end | Integer |
| .startswith() | str.startswith(prefix, start=0, end=len(str)) | prefix | start, end | Boolean |
| .endswith() | str.endswith(suffix, start=0, end=len(str)) | suffix | start, end | Boolean |
| .zfill() | str.zfill(width) | width | None | New string |
| .center() | str.center(width, fillchar=' ') | width | fillchar | New string |
| .format() | str.format(*args, **kwargs) | None | any args | New string |

```python
# zfill,center,format
name="|sam|"

#zfill - is to pad zeroes on the left side
print(name.zfill(5))
print(name.zfill(10))   # first 5 charactor will take as blank

#center - for center alignment
print(name.center(10))
print(name.center(10,'*'))

#format - positional arguments
print("{} was the color of the {} and {} was also used in ther interiors".format("Red","Appples","Red"))
print("{} {} were {} in quantity".format("Red","Apples",10))

#format - keyword arguments
print("{color} {fruit} were {quantity} in quantity".format(quantity=10,color='Yellow',fruit='Mango'))

```
