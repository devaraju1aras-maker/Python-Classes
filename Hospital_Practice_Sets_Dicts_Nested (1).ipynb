## Section 1 — Creating Sets and Removing Duplicates

### Question 1
A hospital recorded blood group entries for walk-in patients. Some patients were recorded twice.

```python
blood_entries = ["A+", "B+", "O+", "A+", "AB+", "B+", "O-", "A+", "B-"]
```

```python
# a. Convert to a set and print unique blood groups
sb=set(blood_entries)
print(sb)
```

```python
# b. How many duplicate entries were there?
print(len(blood_entries)-len(sb))
```

```python
# c. Convert the unique values back to a sorted list
sorted(sb)
```

## Section 3 — Set Operations

### Question 4
Compare morning and evening shift doctors.

```python
morning = {"Dr. Sharma", "Dr. Patel", "Dr. Rao", "Dr. Nair"}
evening = {"Dr. Rao", "Dr. Nair", "Dr. Mehta", "Dr. Khan"}
```

```python
# a. Which doctors worked both shifts?
print(morning.intersection(evening))
```

```python
# b. All doctors who worked today
print(morning.union(evening))
```

```python
# c. Doctors who worked only the morning shift
print(morning.difference(evening))
```

```python
# d. Doctors who worked only the evening shift
print(evening.difference(morning))
```

```python
# e. Doctors who worked exactly one shift
print(morning.symmetric_difference(evening))
```

```python
# f. Solve a, b, c, d using method form instead of symbols
print(morning.intersection(evening))
print(morning.union(evening))
print(morning.difference(evening))
print(evening.difference(morning))
```

### Question 5
Three wards logged their admitted patients.

```python
ward_a = {"P001", "P002", "P003", "P004"}
ward_b = {"P003", "P004", "P005", "P006"}
ward_c = {"P005", "P006", "P007", "P008"}
```

```python
# a. Patients common to ward A and ward B
print(ward_a.intersection(ward_b))
```

```python
# b. Patients common to all three wards
print(ward_a.intersection(ward_b, ward_c))
```

```python
# c. All unique patients across all three wards
print(ward_a.union(ward_b,ward_c))
```

```python
# d. Patients in ward A but not in ward B or C
print(ward_a.difference(ward_b,ward_c))
```

```python
# e. Patients in ward B and C but not in ward A
print(ward_b.intersection(ward_c).difference(ward_a))
```
