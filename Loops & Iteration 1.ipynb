Before writing a loop we need to know what all it needs:
A loop needs keywords,iterable,in keyword ,sequence

An iteartion is anything that can give you one items at a time.
Think of it like a cashier at billing queue,he attends to one customer at a time.

cart["shoes","kite","mac"]

What makes something itearble?
1. It has a defined order or sequence.
2. Python can as it for the next time.
3. It knows when it has run out of items.

| Iterable | Why it works |
|---|---|
| List `[1, 2, 3]` | Has items in order |
| String `"hello"` | Has characters in order |
| Dictionary `{"a": 1}` | Has keys in order |
| Tuple `(1, 2, 3)` | Has items in order |
| Set `{1, 2, 3}` | Has items — no guaranteed order |
| Range `range(5)` | Generates numbers on demand |

```python
#Enumerate - to print index along with the value

c=["A","B","$","%","&"]
for ins,item in enumerate(c):
    print(f"{ins+1}:{item}")
```

Rules and Properties
1. Loops repeat a block of code — the block runs once for each item in a sequence or until a condition becomes False.
2. Indentation is mandatory — everything inside the loop must be indented 4 spaces.
3. Every loop line ends with a colon : — missing it throws a SyntaxError.
4. Two types of loops — for iterates over a sequence. while runs until a condition is False.
5. Loops can be nested — a loop inside a loop.
6. Three control keywords — break exits the loop, continue skips to next iteration, pass does nothing.

#While loops-

A while loop keeps running as long as the condition is true.It does not iterate over a sequence-it checks condition before every single iteration.

syntax:
while condition:
     #code to run while condition is true

A while loop is used when you dont know how many times you need to run or repeat.
It depends on the condition,not a fixed count.

```python
payment_status=input("Enter the payment_status")

while payment_status=="paid":
    print("Enter payment_status")
    break
```

```python
payment_status=input("Enter payment_status")
while payment_status!="paid":
    payment_status=input("Enter payment_status")
    
```

```python
stock=5

while stock>0:
    print(f"Selling stock-stock left:{stock}")
    stock=stock-1

stock=10

while stock>0:
    print(f"Selling stock-stock left:{stock}")
    stock=stock-2
```

```python
#continue

bill_value=[500,9000,0,0,944,4567,4673]

for i in bill_value:
    if i==0:
        continue    #skips the current iteration
    print("The bill value is:",i)
```

```python
#pass - do nothing/placeholder

categories=["Electronics","Grocery","Stationary","Furniture","Electronics"]

for i in categories:
    if i=="Electronics":
        pass
    print("The catagory is:",i)
```

```python
#continue using while  loop

bill_value=[500,9060,0,0,644,4567,73]
index=0
while index<(len(bill_value)):
    if bill_value[index]==0:
       index=index+1
       continue  #skips the current iteration
    print("The bill value is:",bill_value[index])
    index=index+1

#for loop changes index number by default
#while loop not change 
```

```python
#continue using while  loop

bill_value=[500,9060,0,0,644,4567,73]
index=0
while index<(len(bill_value)):
    if bill_value[index]==0:
       index=index+1
       break #skips the current iteration
    print("The bill value is:",bill_value[index])
    index=index+1
```

```python
#List of dictionares 
#skip orders with amount 0 using continue
#stop at cancelled using break
#print only orders that delivered
# expected output ORD001-Deliverd-Rs.2600
orders=[
    {"id":"ORD001","status":"delivered","amount":2600},
    {"id":"ORD002","status":"pending","amount":0},
    {"id":"ORD003","status":"delivered","amount":590},
    {"id":"ORD004","status":"cancelled","amount":9870},
    {"id":"ORD003","status":"delivered","amount":250}
]


for i in orders:
    if i["amount"]==0:
        continue
    elif i["status"]=="cancelled":
        break
    elif i["status"]=="delivered":
        print(f"{i['id']}-{i['status']}-Rs.{i['amount']}")
```
