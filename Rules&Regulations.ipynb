#List comprehension

Rules & Properties:

1. Always creates a new list-the original list is never changed.
2. Square brac[]-same as a regular list.
3. Three parts - expression ,for clause.
4. if-else goes before for when you want to transform every time diiferently.
5. if-only goes after for -when you want to filter items.

Syntax:[expression for varaible in iterable if condition]
        [#ord for odr in orders if odr>1000]
expression - what goes inside the new list .
for variable in iterable- loop over each item
if condition - filter - only include if true

```python
#Add large oders into list
#without list comprehesion
orders=[123,4567,8976,56,7,4321,1000]
large_amounts=[]

for i in orders:
    if i>1000:
        large_amounts.append(i)
print(large_amounts)
```

```python
#iwith list comprehension 
orders=[123,4567,8976,56,7,4321,1002]
large_amounts=[niv for niv in orders if niv>1000]
print(large_amounts)
```

```python
orders=[200,8900,890,8000,5000,80,90,]
discounted=[]

for i in orders:
    discounted.append(i*0.9)
print(discounted)

#with list comprehesion
ors=[i*0.9 for i in orders]
print(ors)
```

```python
#dictionaries in list

orders = [
    {"id": "ORD001", "status": "delivered", "amount": 1500},
    {"id": "ORD002", "status": "pending",   "amount": 800},
    {"id": "ORD003", "status": "delivered", "amount": 2300},
    {"id": "ORD004", "status": "cancelled", "amount": 450},
    {"id": "ORD005", "status": "delivered", "amount": 1100}
]

#create a new list with orders ids of only delivered order
delivered_ids=[i["id"] for i in orders if i["status"]=="delivered"]
print(delivered_ids)

#create a new list with orders amounts of only cancelled orders
cancel=[i["amount"] for i in orders if i["status"]=="cancelled"]
print(cancel)
```

#Dict compreshenion

Rules & Properties:
1. Always creates a new dictionary-original is never changed.
2. Curly{} - same as regulary dictionary.
3. Key-value pair-expression must have a colon separating key and value.
4. Same filter rules as list comprehesion -if goes after the for clause.

SYNTAX :
{key_expression:value_expression for variable in iterable if condition}

key_expression- what becomes the key
value_expression-what becomes the value
for variable in iterable - loop over each item
if condition - filter - only include if true

```python
products={"laptop":45000,"phone":35000,"earbuds":1500,"bottle":1000,"shoes":500}
products_discounted={}

for i,l in products.items():
     products_discounted.update({i:l*0.9}) 
print(products_discounted)


products_discounted={i:l*0.9 for i,l in products.items()}
print(products_discounted)
```

```python
products={"laptop":45000,"phone":35000,"earbuds":1500,"bottle":1000,"shoes":500}

#discount on products whose values is grether than 1000
#using dict comprehension
products_discounted={i:l*.15 for i,l in products.items() if l>1000}
print(products_discounted)

#discount only earbuds
#using dict comprehesnsion
products_discounted={i:l*.12 for i,l in products.items() if i!="earbuds"}
print(products_discounted)
```

```python
# create dictionary from two lists using dict comprehesion
products=["laptop","phone","earbuds","bottle","shoes"]
prices=[450000,30000,1500,1000,500]

product_prices={product:price for product,price  in zip(products,prices)}
print(product_prices)
```

```python
# Dict comprehension with filter
# only products whose price is >20000
product_prices={"mobile":25000,"laptop":85000,"Tablet":45000,"Earbuds":5000}
premium={i:l for i,l in product_prices.items() if l>20000}
print(premium)
```

```python
# list of dictionaries dict comprehension with filter
orders=[
    {"id": "ORD001", "status": "delivered", "amount": 1500},
    {"id": "ORD002", "status": "pending", "amount": 800},
    {"id": "ORD003", "status": "delivered", "amount": 2300}
]
# a dictionary with order id as key and amount of the same order id as value 
# where status="delivered"

delivered={i["id"]:i["amount"] for i in orders if i["status"]=="delivered"}
print(delivered)
```

```python
# Nested Loops - loops within another loop

categories = ["Electronics", "Clothing"]
products = ["Item A", "Item B", "Item C"]

# Electronics - Item A
#               Item B
#               Item C
# Clothing    - Item A
#             - Item B
#             - Item C

for c in categories:
    print("category:", c)
    for p in products:
        print(f"category:", p)
        
              
```

```python
# Nested for with condition
categories = ["Electronics", "Clothing", "Grocery"]
products = ["Laptop", "Shirt", "Rice", "Mobile"]
electronics = ["Laptop", "Mobile"]

# print category with product only if the product is an electronic item

for c in categories:
    for p in products:
        if c=="Electronics" and p in electronics:
            print("category:",c,"product:",p)
```

```python
# Nested loop on list of dictionaries
# output -
# Order: ORD001
#      - shoes
#      - bag
#      - watch

orders=[
    {"id":"ORD001","items":["shoes","bag","watch"]},
     {"id":"ORD002","items":["Laptop","Mouse"]},
     {"id":"ORD003","items":["Shirt","jeans","belt"]}
]

for b in orders:
    print("orders:",b["id"])
    for c in b["items"]:
        print(f"    -{c}")
```

```python
#output -
# January - North: Rs.12000
# January - South: Rs.15000
# January - East: Rs.9000
# February - North: Rs.14000
# February - South: Rs.13000
# March - North: Rs.16000
# March - South: Rs.17000
# March - East : Rs.12000

sales_grid=[
    [12000,15000,9000],  # jan
    [14000,13000,11000], # feb
    [16000,17000,12000]  # mar
]

months=["January","February","March"]
regions=["North","South","East"]

for index,value in enumerate(sales_grid):
    for i,value in enumerate(sales_grid):
        print(f"{months[index]}- {regions[i]}- Rs. {sales_grid[index][i]}")
```

```python
Rows=["Row A","Row B","Row C"]
Columns=["Col 1","Col 2","Col 3"]
grid=[[1,2,3],
      [4,5,6],
      [7,8,9]]

# output-
# Row A — Col 1: 1
# Row A — Col 2: 2
# Row A — Col 3: 3
# Row B — Col 1: 4
# Row B — Col 2: 5
# Row B — Col 3: 6
# Row C — Col 1: 7
# Row C — Col 2: 8
# Row C — Col 3: 9

for index,value in enumerate(grid):
    for i,value in enumerate(grid):
        print(f"{Rows[index]}- {Columns[i]}: {grid[index][i]}")
```
