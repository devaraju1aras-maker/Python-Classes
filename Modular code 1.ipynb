Functions:
Rules and Properties:

A function is a reusable block of code with a name. You define it once and call it whenever you need it.

Rules and Properties
1. Defined with def keyword — followed by the function name and parentheses.
2. Indentation is mandatory — everything inside the function must be indented 4 spaces.
3. Every function definition ends with a colon : — missing it throws a SyntaxError.
4. Functions must be defined before they are called — you cannot call a function before Python has seen its definition.
5. Parameters are optional — a function can have no parameters, one, or many.
6. Return is optional — if no return statement, the function returns None.
7. Function stops at return — any code after return in the same block never runs.


```python
# Functions
bill_amount=2000

def get_delivery(bill_amount):
    if bill_amount>1000:
        return("free delivery")
    else:
        return("Rs.50 will be charged")
get_delivery(500)
```

| Part | What it is |
|---|---|
| `def` | keyword that starts the function definition |
| `get_delivery` | function name — you choose this |
| `(order_amount)` | parameter — input the function receives |
| `:` | mandatory colon |
| indented block | function body — runs when called |
| `return` | sends a value back to wherever the function was called |

# Docstrings

What is a Docstring
A docstring is a string written at the very beginning of a function that explains what it does, what arguments it takes, and what it returns.

"A docstring is a note to yourself and anyone else who reads your code — explaining what the function does without them having to read the whole body."

| Rule | Detail |
|---|---|
| Always first line of function | Right after def line |
| Use triple quotes """ | Even for one line |
| Describe what not how | What does it do, not how it does it |
| Args section | Each parameter name, type, description |
| Returns section | Type and description of return value |

```python
# single line docstring

def order_summary(bill_amount):
    """ this function returns the total, highest and least for the bill_amount entered"""
    total=sum(bill_amount)
    highest=max(bill_amount)
    least=min(bill_amount)
    return total,highest,least

total,highest,least=order_summary([700,600,6000,7000,5600])
print(total)
print(highest)
print(least)
print(order_summary.__doc__) # to access docsring or documentation of the function.
```

```python
# Google style docstring - mostly followed/industry standard
# Numpy style docstring - used in data science or analytics field
```

```python
# Google style doctring-
def create_order(order_id,product,amount,status="pending"):
    """
    creates and displays an order record.

    args:
        order_id(str):unique order identifier
        product(str): Name of the product ordered
        amount(float): order amount in rupees
        status: order status.defaults to "pending"
        
    Returns:
        str:Formatted order summary string
        """
    return f"{order_id} | {product} | Rs.{amount} | {status}"
```

```python
# Numpy style doctring

def order_summary(amounts):
    """
    Calculates summary statistics for a list of order amounts.

    Parameters
    ----------
    amounts : list
        List of order amounts in rupees

    Returns
    -------
    tuple
        A tuple of (total, highest, lowest) order amounts
    """
    return sum(amounts), max(amounts), min(amounts)
```
