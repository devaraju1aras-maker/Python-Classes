Advanced f-string formatting
{} - to use the variable value between printing
.nf - n number of decimal places
< - left- align
> - right- align
^ - center align
, - thousands separator
b - binary
o - octal


```python
# Alignment and padding
name="Aisha"
print(f"name of her was {name:>8}")

print(f"{'name':<15}  {'score':>8}")
print(f"{"Aisha":<15} {98.2:>8.1f}")
print(f"{"San":<15} {76.578:>8.2f}")
print(f"{"Raj":<15} {83.578:>8.3f}")

salary=120000
print(f"{salary:,}")

b=255
print(f"{b:b}")
print(f"{b:o}")

```

# allignment
    Alignemnt controls which side the value sticks to inside a fixed-width space.

    Rule : strings default to left-allign, numbers default to right-align.


```python
print(f"{'left':<10}l")
print(f"{'right':>10}l")
print(f"{'center':^10}l")

print(f"{23:>10}")
print(f"{'hi':<10}")
print(f"{'rd':^10}")

```

```python
# Padding
# Padding is the fill character used to fill the empty space.
# Default is space. You put the fill character before the alignment symbol.


print(f"{'rade':*<10}")
print(f"{'rade':*>10}")
print(f"{'rade':*^10}")

```

```python
# Precision
#Precision controls how many decimal places to show for floats.

pi=3.1415643
print(f"{pi:.0f}")
print(f"{pi:.1f}1")
print(f"{pi:.2f}1")
```

```python
#Padding with numbers has a shorcut
print(f"{7:05}")
print(f"{42:08}")
print(f"{35:07}")
```

```python
#width + precision together
print(f"{pi:10.2f}")    #10 is known as padding & 2f is known as precision

print(f"{pi:010.2f}")
```

Precedence-
precedence decides which operator runs first when multiple operators appear in one expression.

An *expression* is anything that evaluates to a value — 2 + 3, x * y, True and x > 5.

### Precedence Table (High → Low)

| Priority | Operator(s) | Description |
|----------|-------------|-------------|
| 1 | () | Parentheses |
| 2 | ** | Exponentiation |
| 3 | +x -x | Unary sign |
| 4 | * / // % | Multiply / Divide |
| 5 | + - | Add / Subtract |
| 6 | << >> | Bitwise shift |
| 7 | & ^ \| | Bitwise AND/XOR/OR |
| 8 | == != > < >= <= | Comparison |
| 9 | not | Logical NOT |
| 10 | and | Logical AND |
| 11 | or | Logical OR |

*Associativity:* most operators → left to right · ** → right to left (only exception)
