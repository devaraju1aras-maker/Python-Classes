```python
monthly_sales=[2000,30000,5000,7000,6000,8000,9000,2000,7000,9000,4000,5000]

#1st quarter sales (Jan, Feb, March)

#slicing [start:stop:step]
#stop value is always going to give -1 of the stop value 

print(monthly_sales[0:3])

#2nd quarter sales (April, May, June)
print(monthly_sales[3:6])

#3rd quarter sales (July, august, september)
print(monthly_sales[6:9])

#4th quarter sales (October, november, december)
print(monthly_sales[9:12])
print(monthly_sales[9:])

print(monthly_sales[:])  # starts form the first value and ends at the very last value
print(monthly_sales[2:]) # starts from the 2nd month and ends at the last value
print(monthly_sales[:4]) # starts with the 1st value and ends at the 3rd value.

#every second month sales

```

```python
#[start:stop:step]

#step tells how many values to jump

monthly_sales=[2000,30000,5000,7000,6000,8000,9000,12000,10000,9000,4000,15000]

print(monthly_sales[0:12:1])
print(monthly_sales[0:12:3])
print(monthly_sales[0::4])
print(monthly_sales[0:12:6])

# step also decides the direction of reading


# first quarter sales in -Ve indexing - (monthly_sales[start:stop])
print(monthly_sales[-12:-9])

# Second quarter sales
print(monthly_sales[-9:-6])

#third quarter sales
print(monthly_sales[-6:-3])

monthly_sales=[2000, 30000, 5000, 7000, 6000, 8000, 9000, 12000, 10000, 9000, 4000, 15000]
print(monthly_sales[::-1]) # reverse the list
print(monthly_sales[::-2])
print(monthly_sales[::-4])
```

```python
#List Methods
#append
#remove
#extend
#copy
#pop
#sort
#reverse

#append - this method is used for add value in end of the  list 

fruits=['mango', 'apple', 'banana']
fruits.append(['guava'])
print(fruits)

#extend - this method is to add multiple items into the end of the list
fruits=['mango', 'apple', 'banana']
fruits.extend(['peach','strawberry'])
print(fruits)

#removes - this mathod is used for remove the items in the list
fruits=['mango', 'apple','peach', 'banana','peach']
fruits.remove('peach')
print(fruits)


#pop - this also removing the valus by uses of indexing position
fruits=['mango', 'apple', 'banana','sapota']
fruits.pop()  #default behaviour deletes only last value ot item in list 
fruits.pop(2)              #delete the indexing value
print(fruits)

#sort - sort used to ascending or decending  the order of the list

d=['1','20','25','15','45']
d.sort()
print(d) 

d=['1','20','25','15','45']
d.sort(reverse=True)
print(d)
```

```python
# packing and unpacking tuple

# Packing means putting multiple values into a single tuple.

# Unpacking means extracting tuple values into separate variables.

# Rule of Unpacking - Number of variables must match the number of elements.

# Extended Unpacking (*) - The * collects remaining values into a list.

#packing & tuple
a=1,2,3,4,5
print(a)
print(type(a))

t5=("Rahul Verma","ICU",52000,"Dr.Rohan")
#unpacking a tuple
Patient_name,ward,fees,doctor_name=t5

print(Patient_name)
print(ward)
print(fees)
print(doctor_name)

numbers=(123,67,87,546,7654)
start,*middle,end=numbers

print(start)
print(middle)
print(end)

numbers=(123,67,87,546,7654)
*start,end=numbers
print(start)
print(end)

numbers=(123,67,87,546,7654)
_,_,third,fourth,_,=numbers

print(third)
print(fourth)
```

# Hashability
 1. Every value stored inside a set has a hash value.
 2. has value is a randomly generated integer value given to an elements as id
 3. list, dict, set are unhashable items because they are mutable.
 4. unhashable items are not allowed inside a set.
 5. whenever a duplicate value tires to enter the set it is eliminated because of has value being same.

# Set operation.
| Operation | Symbol | Method form | What it does |
|---|---|---|---|
| Union | \| | .union() | All items from both sets |
| Intersection | & | .intersection() | Only items present in both |
| Difference | - | .difference() | Items in one but not the other |
| Symmetric Difference | ^ | .symmetric_difference() | Items in either but not both |

```python
# set operation
c1={"Devaraj","Deepika","Rade","Harshi","Krish","Harshi","Krish"}
c2={"Giri","Hari","Macha","Mama","Bhamaida","Devaraj"}

#Union - all the elements from set a and set b
c3=c1.union(c2)
c4=c1|c2
print(c3)
print(c4)

#intersection - only common elements
c3=c1.intersection(c2)
c4=c1&c2
print(c3)
print(c4)

#difference - present in one but not the other
c3=c1.difference(c2)
c4=c1-c2
print(c3)
print(c4)

#gives elements present in c1 but not in c2



#symmetric difference - elements present in either one of the set but not in 
c3=c2.symmetric_difference(c1)
c4=c2^c1
print(c3)
print(c4)

```
