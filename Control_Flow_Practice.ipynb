## Section 8 — match-case

```python
order_status = "returned"

# 1. Write match-case for:
#    "processing"       → "Order is being packed"
#    "dispatched"       → "Order on its way to you"
#    "delivered"        → "Order delivered"
#    "returned"         → "Return initiated"
#    anything else      → "Status not available"

order_status = input("Enter order status: ")

match order_status:
    case "processing":
        print("Order is being packed")
    case "dispatched":
        print("Order on its way to you")
    case "delivered":
        print("Order delivered")
    case "returned":
        print("Order returned")
    case _:
        print("Unknown status")
```

```python
payment_method = "UPI"

# 2. Write match-case for:
#    "UPI"          → "UPI payment — instant confirmation"
#    "credit card"  → "Credit card — confirmation in 2 hours"
#    "net banking"  → "Net banking — confirmation in 4 hours"
#    "COD"          → "Cash on delivery — pay at doorstep"
#    anything else  → "Payment method not supported"

payment_method = input("Enter payment method: ")

match payment_method:
    case "UPI":
        print("UPI payment — instant confirmation")
    case "credit card":
        print("Credit card — confirmation in 2 hours")
    case "net banking":
        print("Net banking — processing")
    case _:
        print("Unsupported payment method")
```

```python
category = "electronics"

# 3. Write match-case for:
#    "electronics"  → "1 year warranty"
#    "clothing"     → "15 day return"
#    "grocery"      → "No return policy"
#    "furniture"    → "Free installation"
#    anything else  → "Standard policy applies"

category = input("Enter category: ")

match category:
    case "electronics":
        print("1 year warranty")
    case "clothing":
        print("15 day return")
    case "grocery":
        print("No return policy")
    case "furniture":
        print("Free installation")
    case _:
        print("Category not recognized")
```

```python
refund_status = "approved"

# 4. Write match-case for:
#    "approved"     → "Refund will be credited in 5-7 days"
#    "processing"   → "Refund is being processed"
#    "rejected"     → "Refund request rejected. Contact support"
#    "pending"      → "Refund request under review"
#    anything else  → "Unknown refund status"

refund_status = input("Enter refund status: ")

match refund_status:
    case "approved":
        print("Refund will be credited in 5-7 days")
    case "processing":
        print("Refund is being processed")
    case "rejected":
        print("Refund request rejected")
    case _:
        print("Invalid refund status")
```

## Section 9 — Mixed (Stretch)

```python
# 1. E-commerce checkout logic
order_amount   = 1200
payment_status = "paid"
stock          = 3
is_member      = True
coupon_code    = ""

# Step 1 — Check if cart is valid (order_amount > 0) using truthy
# Step 2 — Check if payment is paid
# Step 3 — If paid and stock > 0 — confirm order
#           If paid but no stock — inform customer
# Step 4 — If member — print "Member discount: 10% off"
#           else — print "No member discount"
# Step 5 — Use ternary to assign delivery charge:
#           Free if order >= 1000, else Rs.49
# Step 6 — Print final summary using f-string

order_amount   = 1200
payment_status = "paid"
stock          = 3
is_member      = True
coupon_code    = ""

# Step 1
if order_amount:
    print("Cart is valid")
else:
    print("Cart is empty")

# Step 2 & 3
if payment_status == "paid":
    if stock > 0:
        print("Order confirmed")
    else:
        print("Out of stock")
else:
    print("Payment pending")

# Step 4
if is_member:
    print("Member discount: 10% off")
else:
    print("No member discount")

# Step 5
delivery_charge = "Free" if order_amount >= 1000 else "Rs.49"

# Step 6
print(f"Order Amount: {order_amount} | Payment: {payment_status} | Stock: {stock} | Delivery: {delivery_charge}")
```

```python
# 2. Product review system
rating       = 4.7
review_count = 0
verified     = True

# Step 1 — Check if review_count is non-zero using truthy
#           Print "Reviews available" or "No reviews yet"
# Step 2 — Use if-elif-else to classify rating:
#           >= 4.5 → Excellent, >= 3.5 → Good, >= 2.5 → Average, else → Poor
# Step 3 — Use ternary to assign badge:
#           "Verified Purchase" if verified else "Unverified"
# Step 4 — Print: "[badge] | Rating: [rating] | [classification]"

rating       = 4.7
review_count = 0
verified     = True

# Step 1
if review_count:
    print("Reviews available")
else:
    print("No reviews yet")

# Step 2
if rating >= 4.5:
    classification = "Excellent"
elif rating >= 3.5:
    classification = "Good"
elif rating >= 2.5:
    classification = "Average"
else:
    classification = "Poor"

# Step 3
badge = "Verified Purchase" if verified else "Unverified"

# Step 4
print(f"{badge} | Rating: {rating} | {classification}")
```

```python
# 3. Return request system
order_status = "delivered"
return_days  = 5
reason       = "damaged"

# Step 1 — Use match-case on order_status:
#           Only "delivered" can be returned — print "Return eligible"
#           Any other status — print "Return not applicable" and stop
# Step 2 — If eligible, check return_days:
#           <= 7  → full refund
#           <= 15 → partial refund
#           else  → return window closed
# Step 3 — Use match-case on reason:
#           "damaged"    → "Refund + replacement offered"
#           "wrong item" → "Exchange initiated"
#           "not needed" → "Store credit issued"
#           else         → "Standard return process"

order_status = "delivered"
return_days  = 5
reason       = "damaged"

# Step 1
match order_status:
    case "delivered":
        print("Return eligible")
    case _:
        print("Return not applicable")

# Step 2
if return_days <= 7:
    print("Full refund")
elif return_days <= 15:
    print("Partial refund")
else:
    print("Return window closed")

# Step 3
match reason:
    case "damaged":
        print("Refund + replacement offered")
    case "wrong item":
        print("Exchange initiated")
    case "not needed":
        print("Store credit issued")
    case _:
        print("Standard return process")
```
