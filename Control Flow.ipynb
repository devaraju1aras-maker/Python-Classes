Rules and properties 

1. A condition always evaluates to True or False. python block to run based on this result
2. Indentation is mandatory - the code block under with every condition must be indeneted 4 spaces.
3. Every conditional line ends with : - missing the colon throws the error (Syntax)
4. Only one block runs - in an if-eli-else chain , as soon as one condition is true python runs that blocks and skips everything else.
5. Else is optional - if no condition is matches and there is no else,nothing happens and python moves on
6. Conditions can be combined - using and,or,not.

Truthy or Falsy statements

[]
0
()
{}
""   - above all return false when in blocks statement

Python does not just check True or False.It evealuates any values as truthy or falsy. this means you can use any value directly in a condition  - not just boolean True or false.

```python
#Falsy - these blocks do not run

if 0:
    print("Runs")
if "":
    print("Runs")
if []:
    print("Runs")
if ():
    print("Runs")
if None:
    print("Runs")
if False:
    print("Runs")
```

```python
#Truthy - these are evaluate true

if 1:
    print("Runs")
if "Ansha":
    print("Runs")
if [4,5]:
    print("Runs")
if (3,5):
    print("Runs")
```

```python
#advantage of truthy ot falsy - it could make our condition shorter,simpler 
cart=["mouse","bottle","RT"]

if len(cart)>0:
    print("cart has items")
if cart:
    print("cart has items")    
```

```python
#nested if statement

#if payment is paid then print payment confirmed
#and then check if stock is >0 then print order dispatching
#and then print thank you for shopping

payment_status="Paid"
stock=int(input("Enter the stock value available"))

if payment_status=="Paid":
    print("Payment Confirmed")
    if stock>0:
        print("Order is Dispatching")
    else:
        print("Order is processing")
    print("Thank you for shopping")
```

```python
#match - case statement

#exact match
#multiple inputs that require multiple outputs

#Enter the day name and the day number of the week 
day_of_the_week=input("Enter the day of week")
match day_of_the_week:
    case "Sunday":
        print("First day of the week")
    case "Monday":
        print("Second day of the week")
    case "Tuseday":
        print("Third day of the week")
    case "Wednesday":
        print("Fourth day of the week")
    case "Thursday":
        print("Fifth day of the week")
    case "Friday":
        print("Sixth day of the week")
    case "Saturday":
        print("Seventh day of the week")
    case _:                              #Defaul block when it runs all above conditions is false
        print("Invaild Entry") 
```
